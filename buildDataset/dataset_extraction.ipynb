```python
def get_players_clubV1(url):
    # Transfermarkt requiere un user-agent
    headers = {
        "User-Agent": "Mozilla/5.0"
    }
    

    # Solicitar la página
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")

    # Buscar la tabla de jugadores
    table = soup.find("table", class_="items")

    # Extraer enlaces de los jugadores
    players = []

    if table:
        for row in table.find_all("tr", class_=["odd", "even"]):
            player_cell = row.find("td", class_="hauptlink")
            if player_cell:
                link_tag = player_cell.find("a", href=True)
                name = link_tag.text.strip()
                relative_url = link_tag["href"]
                full_url = f"https://www.transfermarkt.com{relative_url}"
                players.append({"name": name, "url": full_url})
    unique_players = [dict(t) for t in {tuple(player.items()) for player in players}]
    return unique_players
```

```python
def get_players_league(df,lista_total):
    total = len(df)
    for i,url in enumerate(df["url"],start=1):
        print(f"[{i}/{total}] Procesando: {url}")
        lista = get_players_clubV1(url)
        lista_total=lista_total + (lista)
    return lista_total
```

```python
def get_injury_history(name,url):
    injury_url = url.replace("/profil/","/verletzungen/")
    headers = { "User-Agent": "Mozilla/5.0" }
    response = requests.get(injury_url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")


    table = soup.find("table",class_="items")
    if not table:
        print("No se encontro la tabla de lesiones")
        return pd.DataFrame()
    
    data = []
    for row in table.find_all("tr",class_=["odd","even"]):
        cells = row.find_all("td")
        if len(cells)<5:
            continue
        injury = {
            "name": name,
            "season": cells[0].text.strip(),
            "injury": cells[1].text.strip(),
            "from": cells[2].text.strip(),
            "until": cells[3].text.strip(),
            "days_out": cells[4].text.strip(),
            "missed_matches": cells[5].text.strip()
        }
        data.append(injury)

    return pd.DataFrame(data)
```

```python
def name_tokens_in_full_name(name_short, name_full):
    tokens_short = name_short.lower().split()
    tokens_full = name_full.lower().split()
    return all(token in tokens_full for token in tokens_short)
```

```python
def find_short(name_full,short_names):
    for ns in short_names:
        if name_tokens_in_full_name(ns, name_full):
            return ns
    return None
```

```python
import pandas as pd

def substring_merge_players(df1, df2, verbose=True):
    matches = []
    total = len(df1)
    print(total)
    print(len(df2))

    for i, row1 in df1.iterrows():
        name1 = row1["player"]
        season1 = row1["season"]

        if verbose:
            print(f"[{i+1}/{total}] Buscando match para: '{name1}' - Temporada {season1}")

        df2_filtered = df2[df2["season"] == season1] ##Traer los equipos para hacerlo más rapido

        found = False
        for _, row2 in df2_filtered.iterrows():
            name2 = row2["long_name"]
            if name_tokens_in_full_name(name1,name2):
                if verbose:
                    print(f"✅ Match: '{name1}' ⊂ '{name2}'")

                match_row = row1.to_dict()
                match_row.update({f"{k}_df2": v for k, v in row2.items()})
                matches.append(match_row)
                found = True
                break  

        if not found and verbose:
            print(f"❌ Sin match para '{name1}'")

    return pd.DataFrame(matches)
```
