```python
import pandas as pd
```

```python
### Vamos a hacer otro dataset de información del fifa
fifa18 = pd.read_csv("../infoFifa/players_18.csv")
fifa19 = pd.read_csv("../infoFifa/players_19.csv")
fifa20 = pd.read_csv("../infoFifa/players_20.csv")
fifa21 = pd.read_csv("../infoFifa/players_21.csv")
fifa22 = pd.read_csv("../infoFifa/players_22.csv")
fifa23 = pd.read_csv("../infoFifa/players_23.csv")
fifa24 = pd.read_csv("../infoFifa/players_24.csv")
```

```python
ligas = ["English Premier League","Italian Serie A","French Ligue 1", "German 1. Bundesliga","Spain Primera Division"]
fifa18_filtrado = fifa18[fifa18["league_name"].isin(ligas)]
```

```python
columns = ["short_name","long_name","player_positions", "height_cm", "weight_kg", "work_rate", "body_type",
           "pace", "shooting", "passing", "dribbling",	"defending","physic",
           "movement_acceleration",	"movement_sprint_speed","movement_reactions",	"movement_balance", "power_stamina",	"power_strength"
]
```

```python
fifa18_filtrado = fifa18_filtrado[columns]
```

```python
fifa19_filtrado = fifa19[fifa19["league_name"].isin(ligas)]
fifa19_filtrado= fifa19_filtrado[columns]

fifa20_filtrado = fifa20[fifa20["league_name"].isin(ligas)]
fifa20_filtrado= fifa20_filtrado[columns]

fifa21_filtrado = fifa21[fifa21["league_name"].isin(ligas)]
fifa21_filtrado= fifa21_filtrado[columns]

fifa22_filtrado = fifa22[fifa22["league_name"].isin(ligas)]
fifa22_filtrado= fifa22_filtrado[columns]
```

```python
ligas_v2 = ["Premier League","Serie A","Bundesliga","La Liga","Ligue 1"]
fifa23_filtrado = fifa23[fifa23["league_name"].isin(ligas_v2)]
fifa23_filtrado = fifa23_filtrado[columns]
fifa23_filtrado.drop_duplicates(subset=['short_name'],inplace=True)
```

```python
fifa24_filtrado = fifa24[fifa24["league_name"].isin(ligas_v2)]
fifa24_filtrado = fifa24_filtrado[columns]
fifa24_filtrado.drop_duplicates(subset=['short_name'],inplace=True)
```

```python
fifa18_filtrado["season"]="18/19"
fifa19_filtrado["season"]="19/20"
fifa20_filtrado["season"]="20/21"
fifa21_filtrado["season"]="21/22"
fifa22_filtrado["season"]="22/23"
fifa23_filtrado["season"]="23/24"
fifa24_filtrado["season"]="24/25"
```

```python
df_fifa = pd.concat([fifa18_filtrado,fifa19_filtrado, fifa20_filtrado,fifa21_filtrado,
                     fifa22_filtrado,fifa23_filtrado,fifa24_filtrado], axis=0)
```
