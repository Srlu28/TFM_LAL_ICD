```python
import pandas as pd
```

```python
dataframe = pd.read_csv("../datasets/final_df.csv")
```

```python
def define_gravity(df,name,treshold):
    df[name] = df["days_out"].apply(
        lambda x: int(x.split()[0]) >= treshold
    )
    return df
```

```python
dataframe = define_gravity(dataframe,"gravity_injury_90",90)
```

```python
dataframe.head()
```

```python
dataframe.drop(columns=["Unnamed: 0","Unnamed: 0_x"],inplace=True)
```

Este preprocesado consiste en agrupar el dataframe a nivel temporada, es decir para cada jugador y temporada me vas a decir si durante esa temporada sufrio alguna lesion

```python
grouped_df_90 = dataframe.groupby(['player', 'season_x'])['gravity_injury_90'].any().reset_index()

grouped_df_90.rename(columns={
    'season_x': 'season',
    'gravity_injury_90': 'had_major_injury_90'
}, inplace=True)
```

```python
dataframe_aux = dataframe.drop(columns=["injury","from","until","days_out","missed_matches","gravity_injury_90"]).drop_duplicates()
```

```python
dataframe_aux.head()
```

```python
dataframe_study = pd.merge(grouped_df_90,dataframe_aux,how = "left",left_on=['player', 'season'],right_on = ["player","season_x"],suffixes=("_1","_2"))
```

```python
dataframe_study
```

```python
dataframe_study.to_csv("datasetAgrupado90.csv")
```
